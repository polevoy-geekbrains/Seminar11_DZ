Найти производную для нахождения экстремума


```python
from sympy import *
x = symbols('x', real=True)
y = (sin(x)/2 + cos(x)/2) ** 2 * sin(x)
roots = solve((sin(x)/2 + cos(x)/2) ** 2 * sin(x))
der_y = diff(y)
der_y
```

Корни производной (точки экстремума)

```python
roots_der_y = solve(der_y)
roots_der_y
```
